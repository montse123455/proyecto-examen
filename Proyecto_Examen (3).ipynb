### MARCO TEORICO

 Factores Determinantes del Rendimiento Acadmico

El rendimiento acadmico es un fenmeno complejo influenciado por mltiples factores relacionados con el estudiante, su entorno familiar, escolar y socioeconmico. Este marco terico describe y analiza estas variables, destacando su impacto basado en evidencias obtenidas.

 **Factores Personales**
   - **Horas de estudio**: Dedicar tiempo al estudio semanalmente es un determinante clave del aprendizaje, correlacionando positivamente con mejores resultados en exmenes.
   - **Horas de sueo**: Un descanso adecuado, entre 7 y 9 horas por noche, mejora la retencin de informacin y la capacidad cognitiva.
   - **Motivacin**: El nivel de motivacin del estudiante (bajo, medio o elevado) influye en su compromiso con las metas acadmicas. Un nivel alto de motivacin se asocia con mayores logros escolares.

**Entorno Familiar**
   - **Participacin parental**: La implicacin de los padres en la educacin del estudiante (baja, media o alta) est directamente relacionada con un mejor desempeo acadmico, ya que genera apoyo emocional y acadmico.
   - **Ingreso familiar**: Los niveles socioeconmicos (bajo, medio, elevado) impactan el acceso a recursos y oportunidades educativas, influyendo en los resultados de aprendizaje.
   - **Nivel educativo de los padres**: Los padres con mayor nivel educativo (preparatoria, universidad, posgrado) tienden a proporcionar un entorno ms estimulante y con mayores expectativas acadmicas.

**Entorno Escolar**
   - **Calidad del profesorado**: Profesores mejor preparados y con habilidades pedaggicas eficaces influyen significativamente en el rendimiento estudiantil.
   - **Tipo de escuela**: Aunque los anlisis indican que no siempre hay diferencias significativas entre escuelas pblicas y privadas, los recursos disponibles pueden variar.
   - **Influencia de los compaeros**: Una influencia positiva entre los pares puede motivar mejores resultados, mientras que una influencia negativa puede afectar el desempeo.

**Acceso a Recursos**
   - **Acceso a internet y otros recursos educativos**: La disponibilidad de herramientas educativas modernas mejora las oportunidades de aprendizaje.
   - **Actividades extracurriculares**: Participar en actividades extracurriculares no solo enriquece la experiencia educativa, sino que fomenta habilidades adicionales como la disciplina y el trabajo en equipo.

**Aspectos Contextuales**
   - **Distancia a la escuela**: La proximidad influye en la energa y motivacin del estudiante, siendo ms favorable una distancia cercana.
   - **Gnero**: Aunque no siempre se encuentran diferencias significativas, ciertas habilidades o reas de aprendizaje pueden variar por gnero.

Evidencias Estadsticas
El anlisis del archivo confirma:
- Diferencias significativas en puntajes de examen entre grupos basados en participacin parental, nivel de motivacin, y calidad del profesorado.
- No hay diferencias significativas por gnero o tipo de escuela.



**METODOLOGIA**

El enfoque de esta investigacin es cuantitativo, y se adopta un diseo correlacional para examinar las relaciones entre varias variables independientes (participacin parental, acceso a recursos educativos, motivacin acadmica, calidad del profesorado) y la variable dependiente (rendimiento acadmico). A travs de un enfoque correlacional, el estudio busca identificar si existe una relacin significativa entre las variables estudiadas y cmo estas pueden influir en el rendimiento acadmico de los estudiantes.

** Anlisis de datos**

Prueba de Normalidad Para determinar si las variables se ajustan a una distribucin normal, se utilizar el test de Shapiro-Wilk. Si las variables siguen una distribucin normal, se proceder con pruebas paramtricas; de lo contrario, se recurrir a tcnicas no paramtrica.

Anlisis de Varianza (ANOVA) Se aplica la prueba ANOVA de un solo factor para evaluar las diferencias en el rendimiento acadmico segn los diferentes niveles de las variables independientes (participacin parental, acceso a recursos, motivacin y calidad del profesorado).

**Nivel de significancia**

Se utilizar un nivel de significancia de 0.05 ( = 0.05). Esto significa que si el valor p obtenido es menor que 0.05, se rechazar la hiptesis nula y se considerar que existen diferencias significativas entre los grupos.

**Interpretacin de los resultados**

Prueba ANOVA: Si el valor p es menor que 0.05, se concluir que hay diferencias significativas entre las medias de los grupos en trminos de rendimiento acadmico. De ser as, se proceder a realizar la prueba de Tukey para identificar qu grupos especficos presentan diferencias.

Prueba de Tukey: Esta prueba permitir identificar los pares de grupos con diferencias significativas en el rendimiento acadmico, proporcionndonos informacin sobre qu niveles de cada variable son ms influyentes.

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/montse123455/cuarta-unidadd/refs/heads/main/dataset.csv')
df.drop(columns=[], inplace=True)
df.dropna(inplace=True)
df
```

este codigo nos ayuda analizar la variable de la participacion, sus niveles de respuesta son: bajo,medio y elevado.

```python
# Si todos los grupos tienen una distribucin normal
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
    df[df['participacion_parental'] == 'medio']['puntaje_examen'],
    df[df['participacion_parental'] == 'elevado']['puntaje_examen'],
)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadstico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusin: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusin: No hay diferencias significativas entre los tratamientos.")
```

Me diante este codigo podemos ver si que sitiene una fuerte influencia en los examenes

```python
# Si todos los grupos tienen una distribucin normal.
# Si no cumplen con la normalidad, toman el resultado de la prueba de Kruskal
#     y pueden usar los grficos de Tukey para tomar una decisin (ignorando la tabla).
#.    Estrictamente hablando, tendran que realizar la prueba de Dunn.
from statsmodels.stats.multicomp import pairwise_tukeyhsd

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Grfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="nivel_educativo_de_los_padres", xlabel="Diferencia de medias")
```
