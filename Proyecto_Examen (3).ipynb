**INTRODUCCIÓN**


El desempeño académico de los estudiantes es un tema central en el ámbito educativo y social, ya que se reconoce como un indicador clave del desarrollo personal y del progreso de las sociedades. Diversos factores influyen en el rendimiento académico, desde las características individuales de los estudiantes hasta las condiciones socioeconómicas, culturales y educativas que los rodean. Identificar y analizar estos factores permite no solo comprender las dinámicas del aprendizaje, sino también proponer estrategias efectivas para mejorar los resultados en distintos contextos.

Este estudio se enfoca en investigar la influencia de múltiples variables en el rendimiento académico de los estudiantes, incluyendo aspectos como la participación parental, el acceso a recursos, el nivel de motivación, la calidad del profesorado, y factores contextuales como el tipo de escuela, el género, y el ingreso familiar. Estas variables, aunque estudiadas de manera aislada en ocasiones, representan un conjunto complejo de interacciones que moldean las oportunidades de aprendizaje.

El objetivo de este trabajo es identificar patrones y relaciones significativas entre estas variables y el rendimiento académico, utilizando un enfoque cuantitativo y análisis de datos. Al explorar estos factores, se busca no solo entender las causas del éxito o fracaso escolar, sino también proponer recomendaciones para fortalecer los sistemas educativos y promover la equidad en el acceso a oportunidades educativas de calidad.

### MARCO TEORICO

 Factores Determinantes del Rendimiento Académico

El rendimiento académico es un fenómeno complejo influenciado por múltiples factores relacionados con el estudiante, su entorno familiar, escolar y socioeconómico. Este marco teórico describe y analiza estas variables, destacando su impacto basado en evidencias obtenidas.

 **Factores Personales**
   - **Horas de estudio**: Dedicar tiempo al estudio semanalmente es un determinante clave del aprendizaje, correlacionando positivamente con mejores resultados en exámenes.
   - **Horas de sueño**: Un descanso adecuado, entre 7 y 9 horas por noche, mejora la retención de información y la capacidad cognitiva.
   - **Motivación**: El nivel de motivación del estudiante (bajo, medio o elevado) influye en su compromiso con las metas académicas. Un nivel alto de motivación se asocia con mayores logros escolares.

**Entorno Familiar**
   - **Participación parental**: La implicación de los padres en la educación del estudiante (baja, media o alta) está directamente relacionada con un mejor desempeño académico, ya que genera apoyo emocional y académico.
   - **Ingreso familiar**: Los niveles socioeconómicos (bajo, medio, elevado) impactan el acceso a recursos y oportunidades educativas, influyendo en los resultados de aprendizaje.
   - **Nivel educativo de los padres**: Los padres con mayor nivel educativo (preparatoria, universidad, posgrado) tienden a proporcionar un entorno más estimulante y con mayores expectativas académicas.

**Entorno Escolar**
   - **Calidad del profesorado**: Profesores mejor preparados y con habilidades pedagógicas eficaces influyen significativamente en el rendimiento estudiantil.
   - **Tipo de escuela**: Aunque los análisis indican que no siempre hay diferencias significativas entre escuelas públicas y privadas, los recursos disponibles pueden variar.
   - **Influencia de los compañeros**: Una influencia positiva entre los pares puede motivar mejores resultados, mientras que una influencia negativa puede afectar el desempeño.

**Acceso a Recursos**
   - **Acceso a internet y otros recursos educativos**: La disponibilidad de herramientas educativas modernas mejora las oportunidades de aprendizaje.
   - **Actividades extracurriculares**: Participar en actividades extracurriculares no solo enriquece la experiencia educativa, sino que fomenta habilidades adicionales como la disciplina y el trabajo en equipo.

**Aspectos Contextuales**
   - **Distancia a la escuela**: La proximidad influye en la energía y motivación del estudiante, siendo más favorable una distancia cercana.
   - **Género**: Aunque no siempre se encuentran diferencias significativas, ciertas habilidades o áreas de aprendizaje pueden variar por género.

Evidencias Estadísticas
El análisis del archivo confirma:
- Diferencias significativas en puntajes de examen entre grupos basados en participación parental, nivel de motivación, y calidad del profesorado.
- No hay diferencias significativas por género o tipo de escuela.



**METODOLOGIA**

El enfoque de esta investigación es cuantitativo, y se adopta un diseño correlacional para examinar las relaciones entre varias variables independientes (participación parental, acceso a recursos educativos, motivación académica, calidad del profesorado) y la variable dependiente (rendimiento académico). A través de un enfoque correlacional, el estudio busca identificar si existe una relación significativa entre las variables estudiadas y cómo estas pueden influir en el rendimiento académico de los estudiantes.

** Análisis de datos**

Prueba de Normalidad Para determinar si las variables se ajustan a una distribución normal, se utilizará el test de Shapiro-Wilk. Si las variables siguen una distribución normal, se procederá con pruebas paramétricas; de lo contrario, se recurrirá a técnicas no paramétrica.

Análisis de Varianza (ANOVA) Se aplica la prueba ANOVA de un solo factor para evaluar las diferencias en el rendimiento académico según los diferentes niveles de las variables independientes (participación parental, acceso a recursos, motivación y calidad del profesorado).

**Nivel de significancia**

Se utilizará un nivel de significancia de 0.05 (α = 0.05). Esto significa que si el valor p obtenido es menor que 0.05, se rechazará la hipótesis nula y se considerará que existen diferencias significativas entre los grupos.

**Interpretación de los resultados**

Prueba ANOVA: Si el valor p es menor que 0.05, se concluirá que hay diferencias significativas entre las medias de los grupos en términos de rendimiento académico. De ser así, se procederá a realizar la prueba de Tukey para identificar qué grupos específicos presentan diferencias.

Prueba de Tukey: Esta prueba permitirá identificar los pares de grupos con diferencias significativas en el rendimiento académico, proporcionándonos información sobre qué niveles de cada variable son más influyentes.

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/montse123455/cuarta-unidadd/refs/heads/main/dataset.csv')
df.drop(columns=[], inplace=True)
df.dropna(inplace=True)
df
```

```python
df['participacion_parental'].unique()
```

este codigo nos ayuda analizar la variable de la participacion, sus niveles de respuesta son: bajo,medio y elevado.

```python
# Si todos los grupos tienen una distribución normal
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
    df[df['participacion_parental'] == 'medio']['puntaje_examen'],
    df[df['participacion_parental'] == 'elevado']['puntaje_examen'],
)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadístico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusión: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusión: No hay diferencias significativas entre los tratamientos.")
```

Me diante este codigo podemos ver si que sitiene una fuerte influencia en los examenes

```python
# Si todos los grupos tienen una distribución normal.
# Si no cumplen con la normalidad, toman el resultado de la prueba de Kruskal
#     y pueden usar los gráficos de Tukey para tomar una decisión (ignorando la tabla).
#.    Estrictamente hablando, tendrían que realizar la prueba de Dunn.
from statsmodels.stats.multicomp import pairwise_tukeyhsd

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="nivel_educativo_de_los_padres", xlabel="Diferencia de medias")
```

Asimismo, el ingreso familiar influye directamente, ya que familias con mayores recursos pueden proporcionar mejores herramientas y oportunidades educativas. Del mismo modo, el nivel educativo de los padres resulta crucial, ya que aquellos con mayor nivel académico suelen generar un entorno más estimulante y con mayores expectativas para sus hijos.

Sin embargo, el tipo de escuela (pública o privada) no mostró diferencias significativas en los resultados, lo que sugiere que otros factores, como los recursos disponibles y la calidad de la enseñanza, tienen un peso mayor. El acceso a recursos educativos, como internet y materiales de aprendizaje modernos, también favorece el rendimiento. Igualmente, la participación en actividades extracurriculares no solo enriquece la experiencia educativa, sino que fomenta habilidades como la disciplina y el trabajo en equipo, que impactan de manera indirecta en el desempeño escolar.

Por último, factores contextuales como la cercanía a la escuela y un entorno socioemocional positivo también influyen.
